```python
# Drop columns with high missing values and less predictive power
drop_columns_list = ['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'Utilities', 'Street']
training_data.drop(drop_columns_list, axis=1, inplace=True)
testing_data.drop(drop_columns_list, axis=1, inplace=True)
```

```python
# Filling categorical columns with mode and numerical columns with mean
categorical_columns = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                    'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                    'MSZoning', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType']
numerical_columns = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt', 'BsmtFinSF1', 'BsmtFinSF2',
                  'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']
for column in categorical_columns:
    mode_value = training_data[column].mode()[0]
    training_data[column].fillna(mode_value, inplace=True)
    testing_data[column].fillna(mode_value, inplace=True)

for column in numerical_columns:
    mean_value = training_data[column].mean()
    training_data[column].fillna(mean_value, inplace=True)
    testing_data[column].fillna(mean_value, inplace=True)
```

```python
# Feature engineering
training_data['TotalSquareFeet'] = training_data['TotalBsmtSF'] + training_data['1stFlrSF'] + training_data['2ndFlrSF']
testing_data['TotalSquareFeet'] = testing_data['TotalBsmtSF'] + testing_data['1stFlrSF'] + testing_data['2ndFlrSF']
training_data['HouseAge'] = training_data['YrSold'] - training_data['YearBuilt']
testing_data['HouseAge'] = testing_data['YrSold'] - testing_data['YearBuilt']
training_data['IsRemodeled'] = (training_data['YearRemodAdd'] != training_data['YearBuilt']).astype(int)
testing_data['IsRemodeled'] = (testing_data['YearRemodAdd'] != testing_data['YearBuilt']).astype(int)
```

```python
# Drop unused columns
unused_columns = ['YearBuilt', 'YearRemodAdd', 'YrSold', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF']
training_data.drop(unused_columns, axis=1, inplace=True)
testing_data.drop(unused_columns, axis=1, inplace=True)
```

```python
# Encode categorical data
encoder = LabelEncoder()
for column in training_data.columns:
    if training_data[column].dtype == 'object':
        training_data[column] = encoder.fit_transform(training_data[column])
        testing_data[column] = encoder.transform(testing_data[column])
```

```python
# Transform the target variable
training_data['LogSalePrice'] = np.log1p(training_data['SalePrice'])
```

```python
# Prepare data for modeling
features = training_data.drop(['SalePrice', 'LogSalePrice'], axis=1)
target = training_data['LogSalePrice']
X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Model training and evaluation
model_dict = {
    'Random Forest Regressor': RandomForestRegressor(),
    'Linear Regression Model': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Gradient Boosting Regressor': GradientBoostingRegressor()
}

for model_name, model in model_dict.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
    rmse = sqrt(mean_squared_error(y_val, predictions))
    print(f"{model_name} RMSE: {rmse}")
```

```python
# Hyperparameters to tune
param_distributions = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Setup the RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=RandomForestRegressor(),
    param_distributions=param_distributions,
    n_iter=100,
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit RandomizedSearchCV
random_search.fit(X_train, y_train)

# Best model
best_rf = random_search.best_estimator_
```
